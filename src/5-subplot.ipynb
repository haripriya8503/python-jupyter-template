### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x, y, marker='o', color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y Values')
axs[0, 0].grid(True)

axs[0, 1].bar(labels, sizes, color=colors)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Languages')
axs[0, 1].set_ylabel('Usage')

axs[1, 0].scatter(x, y, color='green', marker='o')
for i in range(len(x)):
    axs[1, 0].text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=10, ha='right', va='bottom')
axs[1, 0].set_title('Scatter Plot with Annotations')
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')
axs[1, 0].grid(True)

axs[1, 1].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')

plt.tight_layout()
plt.show()
```
